```python
price=content.find_all('div',{'class':'lfloat marR10'})
#print(price)
#price=price[0].text
len(price)
```

```python
rating=content.find_all('div',{'class':'rating-stars'})
#rating=rating[0].text
#print(rating)
len(rating)
```

```python
discount=content.find_all('div',{'class':'product-discount'})
#rating=rating[0].text
#print(rating)
len(discount)
```

```python
pr=[]
rt=[]
dis=[]
```

```python
for i in price:
    pr.append(i.text)
    
for i in discount:
    dis.append(i.text)
    
for i in range(len(pr)):
    rt.append(rating[i].text)
```

```python
data={'PRICE':pr,'RATING':rt,'Dicount':dis}
df=pd.DataFrame(data)
```
